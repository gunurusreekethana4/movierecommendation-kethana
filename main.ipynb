```python
movies_data.dropna(subset=['title','release_date'],inplace=True)
```

```python
movies_data.drop(columns=['homepage', 'tagline', 'keywords'], inplace=True)
```

```python
movies_data['overview'].fillna('No overview available', inplace=True)
```

```python
movies_data['decade'] = (movies_data['release_date'].str[:4].astype(float) // 10 * 10).astype('Int64')
movies_data['genres'] = movies_data.groupby('decade')['genres'].transform(lambda x: x.fillna(x.mode()[0] if not x.mode().empty else 'Unknown'))
```

```python
movies_data['backdrop_path'].fillna('default_backdrop.jpg', inplace=True)
movies_data['poster_path'].fillna('default_poster.jpg', inplace=True)
```

```python
movies_data.drop('imdb_id',axis=1,inplace=True)
```

```python
movies_data.drop('original_title',axis=1,inplace=True)
```

```python
movies_data.drop(['revenue', 'budget','production_companies','production_countries','decade'], axis=1, inplace=True)
```

```python
movies_data.drop('spoken_languages',axis=1,inplace=True)
```

```python
#feature selection
movies=movies_data[['id','title','overview','genres']]
```

```python
movies
```

```python
movies = movies.copy()  # Create a copy of the DataFrame
movies['tags'] = movies['overview'] + movies['genres']
```

```python
movies
```

```python
new_data = movies.drop(columns=['overview','genres'])
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

vectorizer = TfidfVectorizer(max_features=10000,stop_words='english')
vector = vectorizer.fit_transform(new_data['tags'].values.astype('U'))
print(vector.shape)
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
# Use only the first 10,000 items for testing
subset_vector = vector[:20000]
similarity = cosine_similarity(subset_vector)
```

```python
#  input
#movie_title='Iron Man'
movie_title = input("Enter the movie title: ").strip().lower()  # Convert input to lowercase

# Call the recommendation function
try:
    # Convert all movie titles in the dataset to lowercase for case-insensitive comparison
    new_data['title_lower'] = new_data['title'].str.strip().str.lower()

    # Find the index of the input movie (case-insensitive)
    movie_index = new_data[new_data['title_lower'] == movie_title].index[0]

    # Get similarity scores for the input movie
    sim_scores = list(enumerate(similarity[movie_index]))

    # Sort by similarity score
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)

    # Exclude the input movie itself
    sim_scores = sim_scores[1:6]  # Top 5 recommendations

    # Get movie indices
    movie_indices = [i[0] for i in sim_scores]

    # Return top N recommendations (retain original case)
    recommendations = new_data['title'].iloc[movie_indices]

    # Print the recommendations
    print(f"\nTop 5 recommendations for '{new_data['title'].iloc[movie_index]}':")
    for i, title in enumerate(recommendations, start=1):
        print(f"{i}. {title}")
except IndexError:
    print(f"Error: The movie '{movie_title.title()}' was not found in the dataset.")
```
